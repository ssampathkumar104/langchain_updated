```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
# Creating an agent
from langchain.agents import create_agent

def get_whether(city: str):
    """Get today's weather of a city"""
    return "Today's weather in {city} is sunny."

agent = create_agent(
    model="gpt-3.5-turbo",
    tools=[get_whether],
    system_prompt="You are a helpful AI assistant."
    # verbose=True,
)
agent
```

```python
agent.invoke({"messages": [{"role": "user", "content": "What's the weather like in Delhi?"}]})
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What's the weather like in Delhi?"}]})
response
```
